```python
import os
from math_datasets.datasets import SVAMP, GSM8K

gemini_model_names = [
    # "gemini-2.0-flash",
    "gemma-3-27b-it",
]
```

```python
from math_datasets.evaluator import evaluate_detail
import pandas as pd


df_gsm8k = evaluate_detail("gemma-3-27b-it", GSM8K, os.getcwd(), use_transformated_answers=False)
df_svamp = evaluate_detail("gemma-3-27b-it", SVAMP, os.getcwd(), use_transformated_answers=False)

df = pd.concat([df_gsm8k, df_svamp], ignore_index=True)
print(len(df), len(df_gsm8k), len(df_svamp))
df.head()
```

```python
df = df[df["is_correct"] == True].copy()
print(len(df))
df.head()
```

```python
df.loc[:, "sft_label"] = df["model_history"].apply(
    lambda x: x[0]["plan"]["plan_string"]
)
df.head()
```

```python
prompt = """For the following task, make a detailed step-by-step plan to solve the problem.  
For each step, choose **one tool** to retrieve or calculate the necessary information.  
The tool output should be stored in a variable like #E1, #E2, etc., which can be used in later steps.

Tools available:
(1) LLM[input]: A pretrained language model like yourself. Use this to reason through parts of the problem using general knowledge or logic. Input can be a natural language instruction.
(2) Calculator[input]: Use this for basic arithmetic operations (e.g., addition, subtraction, multiplication, division). Input must be a valid math expression like "290 / 2".

**Do not solve the problem directly. Only write the plan and tool inputs.**  
Each step must follow this format:
Plan: [describe the reasoning for the step] #EX = Tool[tool input]

---

Begin!  
Describe your plans with rich details. Each Plan must be followed by exactly one #E.

Task: {task}"""

df = df[["question", "sft_label"]]
df = df.rename(columns={"sft_label": "output"})
df["input"] = df["question"].apply(
    lambda x: prompt.format(task=x)
)
df = df[["question", "input", "output"]]
df.head()
```

```python
print("Input:\n" +df.iloc[0]["input"])
print("Output:\n" + df.iloc[0]["output"])
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df.to_csv("train_data.csv", index=False)
```

```python
def format_chat(example):
    user_message = example["input"]
    assistant_message = example["output"]
    messages = [
        {"role": "user", "content": user_message},
        {"role": "assistant", "content": assistant_message}
    ]
    return messages

df["messages"] = df.apply(format_chat, axis=1)
df.to_json("rewoo_train_data.jsonl", orient="records", lines=True)
```
